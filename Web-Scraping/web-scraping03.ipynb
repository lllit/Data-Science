```python
import pandas as pd
import time
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://dockerlabs.es/'

respuesta = requests.get(url)
```

```python
if respuesta.status_code == 200:
    
    soup = BeautifulSoup(respuesta.text, 'html.parser')

    maquinas = soup.find_all('div', onclick = True)

    conteo_maquinas = 1

    datos = []
    
    # Función para limpiar y decodificar texto con errores de codificación
    def limpiar_texto(texto):
        # Corrige errores de codificación típicos
        return texto.encode('latin1').decode('utf-8')
    
    for maquina in maquinas:
        onclick_text = maquina['onclick']

        # Extraer los datos de la cadena 'onclick'
        nombre = onclick_text.split("'")[1]
        dificultad = onclick_text.split("'")[3]
        autor = onclick_text.split("'")[7]

        # Limpiar los textos antes de añadirlos al DataFrame
        nombre = limpiar_texto(nombre)
        dificultad = limpiar_texto(dificultad)
        autor = limpiar_texto(autor)

        # Almacenar los datos en un diccionario
        datos.append({
            'nombre': nombre,
            'dificultad': dificultad,
            'autor': autor
        })

    # Crear el DataFrame con los datos recopilados
    df = pd.DataFrame(datos)
        
    df.to_csv('../data/datos_maquinas.csv', index=False)
    
else:
    print(f"Error al hacer la petición {respuesta.status_code}")
```
